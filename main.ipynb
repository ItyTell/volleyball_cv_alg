```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.patches as patches
```

```python
# Load the image
img = cv.imread('dataset\\videos_sample\\7\\51725\\51705.jpg', cv.IMREAD_COLOR)
```

```python
# reveal the image

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
# read txt file
df = pd.read_csv('dataset\\volleyball_tracking_annotation\\7\\51725\\51725.txt', sep=' ')
headers = [
    '{Frame ID}', 
    'X',
    'Y',
    'X1',
    'Y1',
    '{Frame Activity Class}', 
    'smth1',
    'smth2',
    'smth3',
    'action',
]
df.drop(df.columns[-1], axis=1, inplace=True)
df.columns = headers
df
```

```python
df1 = df[199:200]
df1
```

```python
df1['X'].values[0]
```

```python
rect = (df1['X'].values[0], df1['Y'].values[0])
rect1 = (df1['X1'].values[0] - rect[0], df1['Y1'].values[0] - rect[1])
rect
```

```python
rect1
```

```python
fig, ax = plt.subplots()

# Display the image
ax.imshow(img)

# Create a Rectangle patch
rect = patches.Rectangle(rect, rect1[0], rect1[1], linewidth=1, edgecolor='r', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
